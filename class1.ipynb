```python
'''
numpy里的基本对象
ndarray  处理单一数据的多维数组
ufunc   能够对数组进行处理的函数
'''
import numpy as np
```

```python
'''
ndarray的属性
'''
# 1.shape 用于表达数组的大小(也称之为维度)
# 2.ndim 用于表达数组的维数
# 3.size 用于表达数组中元素的总数
# 4.dtype 用于表达数组中元素的数据类型
# 5.itemsize 用于表达数组中每个元素的字节大小
```

```python
'''
array函数 创建一个数组的函数
'''
object = [1, 2, 3, 4, 5]
a = np.array(object, dtype=None, copy=True, order='K', subok=False, ndmin=0, like=None)
#参数解释如下：
# object(重要)：表示输入数据，可以是列表、元组、数组等。
# dtype(重要)：表示数组元素的数据类型，可以是int、float、str等。
# copy：表示是否复制输入数据，默认为True。
# order：表示数组的存储顺序，可以是'C'（行优先）或'F'（列优先）。K表示保持输入数据的存储顺序。
# subok：表示是否允许子类数组，默认为False。
# ndmin(重要)：表示数组的最小维数，如果输入数据的维数小于ndmin，则会在前面添加维度。
# like：表示参考对象，如果提供了like参数，生成的数组将具有与参考对象相同的属性。

#后面的参数都没多重要 最重要的就是object和dtype和ndim就OK
```
